Separation of target and feature variables

```python
y = df['Class']
X = df.drop(['Class'], axis =1)
```

```python
 X = df.drop(['Time'], axis =1 )
 df = df.drop(['Time'], axis =1 )
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
print(scaler.fit(X[['Amount']]))
StandardScaler()
```

```python
print(scaler.mean_)
```

```python
print(scaler.transform(X[['Amount']]))
df['Amount'] = scaler.transform(X[['Amount']])
```

Eliminate skewness from data

```python
df.skew()
```

```python
X1 = X.columns
X1
```

```python
for column in df.columns:
  print(column)
```

```python
for column in df.columns:
  a = df[column].skew()
  if a<-1 or a>1:
    df[column] = np.cbrt(df[column])
    print('new_skewness'+ str(column), df[column].skew())
```

```python
# Splitting Data into Training and Test Set
# Using ScikitLearn Library

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 19, stratify = df['Class'])
```

```python
# Implementing LogisticRegression Model

from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
y_train_pred = model.predict(X_train)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_train, y_train_pred)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred ))
```
